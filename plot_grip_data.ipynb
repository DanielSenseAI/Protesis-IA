```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotData(file_path):
    # Read the Excel file
    data = pd.read_excel(file_path)

    # Extract the columns s1 to s8
    s_columns = data[['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8']]

    # Plot the columns
    plt.figure(figsize=(10, 6))
    for column in s_columns:
        plt.plot(s_columns[column], label=column)

    # Adding labels and title
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.title('Plot of s1 to s8 Columns')
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()
```
